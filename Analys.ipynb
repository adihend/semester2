```python
fig = plt.figure()
ax = fig.add_axes ([0,0,3,1])
x = df.Country
y = df.Status
ax.bar(x,y)
plt.xticks (rotation = 'vertical')
plt.show()
```

```python
print("STATUS")
data = pd.read_csv('data_aqi.csv')
counts = data['Status'].value_counts()
total = counts.sum()
percentages = counts / total * 100
labels = [f'{count} ({val})' for val, count in zip(data['Status'].unique(), counts)]
fig, ax = plt.subplots()
ax.pie(counts, labels=labels, autopct='%1.1f%%', startangle=900)
ax.axis('equal')
ax.set_title('Status AQI')
plt.show()
```
